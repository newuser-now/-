```python
# @title 🎵 粤越歌声转换处理器 (粤语专用版)

!pip install -q demucs faster-whisper deep-translator gtts
!apt-get install -qq ffmpeg

import torch
import torchaudio
import subprocess
import gc
from demucs.pretrained import get_model
from demucs.apply import apply_model
from faster_whisper import WhisperModel
from deep_translator import GoogleTranslator
from gtts import gTTS
from google.colab import files

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🚀 GPU: {device}")

uploaded = files.upload()
input_file = list(uploaded.keys())[0]

# ── 1. 分离人声 (CPU) ──────────────────────────────────
print("🎤 分离人声 (CPU)...")
demucs = get_model('htdemucs').to("cpu")
wav, _ = torchaudio.load(input_file)
wav = wav[:2]
sources = apply_model(demucs, wav[None], device="cpu")[0]
torchaudio.save("vocals.wav", sources[3], 44100)
torchaudio.save("backing.wav", sources[:3].sum(0), 44100)

del demucs, wav, sources
gc.collect()

# ── 2. 识别粤语 (专用模型) ─────────────────────────────
print("🗣️ 识别粤语 (专用模型)...")
# 使用支持粤语的 Whisper 模型
asr_model = WhisperModel("medium", device=device, compute_type="float16")
segments, info = asr_model.transcribe("vocals.wav", language="yue")
cantonese = " ".join([segment.text for segment in segments]).strip()
print(f"🇭🇰 粤语: {cantonese}")

del asr_model
torch.cuda.empty_cache()
gc.collect()

# ── 3. 翻译 ────────────────────────────────────────────
print("🌐 翻译...")
vietnamese = GoogleTranslator(source='zh-CN', target='vi').translate(cantonese)
print(f"🇻🇳 {vietnamese}")

# ── 4. 合成语音 (gTTS) ─────────────────────────────────
print("🎹 合成...")
tts = gTTS(text=vietnamese, lang='vi', slow=False)
tts.save("new_vocals.mp3")

subprocess.run([
    'ffmpeg', '-y', '-i', 'new_vocals.mp3',
    '-ar', '44100', '-ac', '1', 'new_vocals.wav'
], check=True, capture_output=True)

# ── 5. 混音 ────────────────────────────────────────────
print("🎚️ 混音...")
subprocess.run([
    'ffmpeg', '-y', '-i', 'backing.wav', '-i', 'new_vocals.wav',
    '-filter_complex', '[1:a]volume=0.8[v];[0:a][v]amix=inputs=2:duration=longest',
    '-q:a', '2', 'output.mp3'
], check=True, capture_output=True)

print("🎉 完成！")
files.download("output.mp3")
```
